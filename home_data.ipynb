# Linear Regression

Let's assume that we desire to sell our house and we would like to predict its price according to its feet square. We would like to build a linear regression model.
Firstly, we should seperate the two features of the house dataset that we will use as coordinates.

```python
houses_coordinates = houses[['price', 'sqft_living']].copy()
houses_coordinates
```

```python
hats = houses_coordinates.mean(axis = 0)
```

```python
hats
```

```python
temp_price = houses_coordinates['price'].add((-1*hats['price']))
temp_sf =  houses_coordinates['sqft_living'].add((-1*hats['sqft_living']))
```

```python
temp_price
```

```python
temp_sf
```

```python
temp_price = temp_price.tolist()
temp_sf = temp_sf.tolist()
```

```python
#Construct functions to find m
def square_sum (temp_sf):
    summ = 0
    for sf in temp_sf:
        summ = summ + sf * sf
    return summ   

def sum_between(temp_sf,temp_price):
    summ = 0
    if len(temp_sf) != len(temp_price):
        return
    else:
        for i in range(0,len(temp_sf)):
            summ = summ + (temp_sf[i] * temp_price[i])
    return summ
    
```

Our aim is to find a line y = mx + c that will fit the data best. 

```python
m = sum_between(temp_sf,temp_price)/square_sum(temp_sf)
m
```

Next step is to find a random point that exists in the dataset and produce a straight line that is a possible solution to our problem.

```python
import random
#Find random point 
def random_choice(houses_coordinates):
    temp = houses_coordinates.to_numpy()
    index = random.randint(0, len(houses_coordinates))
    return temp[index]

point = random_choice(houses_coordinates)
```

```python
c = m*point[1] + point[0]
```

```python
c
```

```python
import numpy as np
x = np.linspace(-1,homes['sqft_living'].max(),2000)
```

```python
homes['sqft_living'].min()
```

```python
y = m*x + c
plt.plot(x, y,'c')
plt.grid()
plt.show()
```

```python
plt.scatter(homes['sqft_living'], homes['price'],alpha=0.5)
plt.plot(x, y,'c')
plt.xlabel('square feet')
plt.ylabel('prices')
plt.show()
```

This is a possible line that might fit the data, but as we can see from the visual presentation above, it might not be the best solution. Now let's try to predict the price of a house given its square feet.

```python
def predict (homes,sqfeet,m):
    point = random_choice(homes)
    c = m*point[1] + point[0]
    y = m*sqfeet + c
    return y
    
```

```python
priceSuggested = predict(houses_coordinates,2000,m)
```

```python
print(priceSuggested)
```
