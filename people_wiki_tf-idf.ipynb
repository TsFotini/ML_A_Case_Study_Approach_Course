# Term Frequency

```python
def tf(doc,term,allCount):
    word_list = nltk.word_tokenize(doc)
    counter = word_list.count(term)
    tuples = (term,float(counter/allCount))
    return tuples
```

```python
def allCount(textlist):
    counter = 0
    for i in textlist:
        word_list = nltk.word_tokenize(i)
        counter = counter + len(word_list)
    return counter
```

```python
allCount = allCount(textlist)
```

```python
elvis_text = df_temp['text'].tolist()
```

```python
elvis_text[0]
```

```python
elvis_list = nltk.word_tokenize(elvis_text[0])
results = []
for i in elvis_list:
    score = tf(elvis_text[0],i,allCount)
    results.append(score)
results
```

# IDFs

```python
def findalldocuments(textlist):
    return len(textlist)
```

```python
def termInDocument(textlist,term):
    counter = 0
    for i in textlist:
        if term in i:
            counter = counter + 1
    return counter
```

```python
import math

def idf (textlist,term,N):
    num = math.log(N/termInDocument(textlist,term),2)
    tuples = (term,num)
    return tuples
```

```python
N = findalldocuments(textlist)
results_idf = []
for term in elvis_list:
    scores = idf(textlist,term,N)
    results_idf.append(scores)
results_idf    
```

# TF-IDF
This is a combination of TF and IDF.

```python
def tf_idfs (tf,idf):
    result = []
    for i in tf:
        for j in idf:
            multy = i[1] * j[1]
        tuples = (i[0],multy)
        result.append(tuples)
    return result

final = tf_idfs(results,results_idf)           
```
