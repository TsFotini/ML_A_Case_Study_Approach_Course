# Forming the matrix

We will start by creating the sparce matrix that contains users and songs.

```python
len(df['song_id'].unique())
```

What we know, is that we have a dataset that contains 66346 users and 10000 songs. The songs have been rated from the users. In this case we have all that we need to create the matrix. 

```python
ratings_list = []
columns_songs = df['song_id'].unique()
index_users = df['user_id'].unique()
matrix_df = pd.DataFrame (ratings_list, index = index_users ,columns = columns_songs)
print (matrix_df)
```

```python
index_users = index_users[0:10]
```

```python
columns_songs = columns_songs[10:40]
```

```python
data_fill = []
for i in index_users:
    datarow = []
    for j in columns_songs:
        df_temp = df.query('`user_id` == @i and `song_id` == @j ')
        value = df_temp['listen_count'].values
        value = value.tolist()
        if len(value)==0:
            value.append(-1)
        datarow.append(value[0])
    data_fill.append(datarow)
```

```python
matrix_df = pd.DataFrame (data_fill, index = index_users ,columns = columns_songs)
```
